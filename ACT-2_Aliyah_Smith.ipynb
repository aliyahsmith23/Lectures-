<h1 style="text-align: center;"> Doing Queries with DataFrame </h1>

```python
# Problem 1

n_row = range(data.shape[0])
columns_list = ["PTGENDER", "DX_bl"]
male = pd.DataFrame(data, index=n_row, columns=columns_list).query("PTGENDER =='Female' and DX_bl == 'AD' ").count()

print("Total number of Males with AD: ", male)
```

```python
# Problem 2
n_row = range(data.shape[0])
columns_list = ["PTGENDER", "DX_bl"]
female = pd.DataFrame(data, index=n_row, columns=columns_list).query("PTGENDER =='Female' and DX_bl == 'AD' ").count()

print("Total number of Females with AD: ", female)
```

```python
# Problem 3 
male_ad_avg_age = data[(data['PTGENDER'] == 'Male') & (data['DX_bl'] == 'AD')]['AGE'].mean()
print("The average of males with AD: ", male_ad_avg_age)
```

```python
# Problem 4

female_ad_avg_age = data[(data['PTGENDER'] == 'Male') & (data['DX_bl'] == 'AD')]['AGE'].mean()
print("The average of females with AD: ", female_ad_avg_age)
```

```python
# Problem 5

n_row = range(data.shape[0])
columns_list = ["PTGENDER", "DX_bl"]
male = pd.DataFrame(data, index=n_row, columns=columns_list).query("PTGENDER =='Female' and DX_bl == 'CN' ").count()

print("Total number of Males with CN: ", male)

male_ad_avg_age = data[(data['PTGENDER'] == 'Male') & (data['DX_bl'] == 'CN')]['AGE'].mean()
print("The average of males with CN: ", male_ad_avg_age)

print("")
n_row = range(data.shape[0])
columns_list = ["PTGENDER", "DX_bl"]
female = pd.DataFrame(data, index=n_row, columns=columns_list).query("PTGENDER =='Female' and DX_bl == 'CN' ").count()

print("Total number of Females with CN: ", female)

female_ad_avg_age = data[(data['PTGENDER'] == 'Male') & (data['DX_bl'] == 'CN')]['AGE'].mean()
print("The average of females with CN: ", female_ad_avg_age)



```
